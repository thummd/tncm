```python
import networkx as nx
import matplotlib.pyplot as plt

# Example graph: Replace this with your actual graph building code
G = nx.DiGraph()
edges = relation[["source","target"]].values
G.add_edges_from(edges)

plt.figure(figsize=(12, 8))

# Use spring layout for neat spacing
pos = nx.spring_layout(G, seed=42)

# Draw nodes and edges
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=50)
nx.draw_networkx_edges(G, pos, arrowsize=10)
nx.draw_networkx_labels(G, pos)

plt.title("Neater Directed Graph")
plt.axis('off')
plt.show()
```
